## 0-2. カメラで 1 枚撮ってみる
Raspberry Pi AI カメラで撮影し、**この画面にそのまま表示**します。
（初回はカメラの初期化に数秒かかります）

```python
import matplotlib.pyplot as plt
%matplotlib inline
from picamera2 import Picamera2
import time

picam = Picamera2()
picam.configure(picam.create_preview_configuration(
    main={'size': (640, 480), 'format': 'RGB888'}))
picam.start(); time.sleep(1)
frame = picam.capture_array()   # OpenCV 互換の BGR 配列
picam.close()

# 表示は RGB に直してから（matplotlib は RGB 前提）
plt.figure(figsize=(6, 4))
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.axis('off'); plt.title('はじめての1枚'); plt.show()
print('画像の形 shape =', frame.shape)
```

✅ **チェックポイント**：カメラの映像が表示され、`shape = (480, 640, 3)` と出れば成功です。

うまく映らないときは、カメラのケーブルの向きを確認（電源を抜いてから挿し直す）。
